```python
import pandas as pd
import random


df = pd.read_csv("datasets/medata_solicitudes_de_credito_analizadas.csv", sep=";")
df_sample1 = df.sample(frac=0.05)
df_sample2 = df.sample(frac=0.01)
df_sample3 = df.sample(frac=0.01)
df_sample4 = df.sample(frac=0.01)
df_sample2['tipo_de_emprendimiento'] = pd.NA
df_sample3['barrio'] = pd.NA
df_sample4['linea_credito'] = pd.NA

df = pd.concat([df, df_sample1, df_sample2, df_sample3])
df = df.sample(frac=1).reset_index(drop=True)

for col in ['sexo', 'tipo_de_emprendimiento', 'idea_negocio', 'barrio', 'línea_credito']:
    sample = random.sample(range(0, 11145), random.randint(100, 500))
    df.loc[sample, col] = df.loc[sample, col].str.upper()
    
    sample = random.sample(range(0, 11145), random.randint(100, 500))
    df.loc[sample, col] = df.loc[sample, col].str.title()
    
    sample = random.sample(range(0, 11145), random.randint(100, 500))
    df.loc[sample, col] = df.loc[sample, col].str.replace(' ', '-')
    
    sample = random.sample(range(0, 11145), random.randint(100, 500))
    df.loc[sample, col] = df.loc[sample, col].str.replace(' ', '_')
    
    
sample = random.sample(range(0, 11145), random.randint(300, 580))
df.loc[sample, 'estrato'] = df.loc[sample, 'estrato'].map(lambda x: '0' + str(x))


sample = random.sample(range(0, 11145), random.randint(150, 400))
df.loc[sample, 'fecha_de_beneficio'] = df.loc[sample, 'fecha_de_beneficio'].map(lambda x: x[-4:] + "/" + x[-7:-5] + '/' + x[-len(x):-8])

sample = random.sample(range(0, 11145), random.randint(100, 500))
df.loc[sample, 'monto_del_credito'] = df.loc[sample, 'monto_del_credito'].map(lambda x: '$ {:0,.2f}'.format(x))




df.to_csv('datasets/clean_solicitudes_de_credito_analizadas.csv', sep=";")
```
