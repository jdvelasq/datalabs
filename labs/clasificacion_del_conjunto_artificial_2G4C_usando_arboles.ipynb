# Clasificación del conjunto artificial de datos de dos grupos y cuatro clusters usando árboles de clasificación

El siguiente es un conjunto artificial de datos que tiene una frontera de decisión compleja. 

```python

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

NPOINTS = 400

plt.figure(figsize=(8,8))

X, y = make_blobs(
    n_samples=NPOINTS, 
    n_features=2,
    centers=[[-3, -3],
             [-3,  3],
             [ 3, -3],
             [ 3,  3]],
    cluster_std=1.0, 
    shuffle=False,  
    random_state=1234)

y[y==3] = 0
y[y==2] = 1

plt.scatter(X[y==0, 0], X[y==0, 1], color='red')
plt.scatter(X[y==1, 0], X[y==1, 1], color='blue')
plt.gca().set_aspect('equal', adjustable='box')
```

Construya un árbol de decisión que permita su clasificación.
